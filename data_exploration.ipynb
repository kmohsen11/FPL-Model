```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

```

```python
df = pd.read_csv('data/final_dataset.csv')
```

```python

# Display basic descriptive statistics
print("Basic Descriptive Statistics:")
print(df.describe())

# Display the count of missing values for each column
print("\nMissing Values in the Dataset:")
print(df.isnull().sum())

# Display unique values for categorical fields (like position)
print("\nUnique Positions in the Dataset:")
print(df['position'].unique())

# Display basic info to check data types and non-null counts
print("\nData Information:")
print(df.info())
```

```python
import matplotlib.pyplot as plt

# Bar plot for the count of player positions
plt.figure(figsize=(10, 6))
df['position'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Player Positions')
plt.xlabel('Position')
plt.ylabel('Count')
plt.show()
```

```python
# Histogram for points in the last 5 games
plt.figure(figsize=(10, 6))
df['points in last 5 games'].plot(kind='hist', bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Points in Last 5 Games')
plt.xlabel('Points')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot distribution of points in last 5 games by position
plt.figure(figsize=(12, 6))
sns.boxplot(x='position', y='points in last 5 games', data=df)
plt.title('Points in Last 5 Games by Position')
plt.xlabel('Position')
plt.ylabel('Points in Last 5 Games')
```

```python
import seaborn as sns
import numpy as np

# Calculate correlations with total_points
correlation_with_points = df.corr()['total_points'].sort_values(ascending=False)

# Display the top correlations
print("Top Correlations with Total Points:")
print(correlation_with_points)

# Plot the top correlations
plt.figure(figsize=(12, 8))
correlation_with_points.drop('total_points').plot(kind='bar', color='teal')
plt.title('Correlation of Features with Total Points (Current Round)')
plt.xlabel('Features')
plt.ylabel('Correlation Coefficient')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()
```

```python
# Plot the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
